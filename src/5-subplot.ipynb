### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = x**2

bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart
axs[0, 0].plot(x, y, color='blue', label="y = x^2")
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values")
axs[0, 0].grid(True)
axs[0, 0].legend()

# Bar chart
axs[0, 1].bar(bar_categories, bar_values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(True, axis='y', linestyle="--", alpha=0.6)

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple', label="Data Points")
for i in range(len(scatter_x)):
    axs[1, 0].annotate(f"({scatter_x[i]}, {scatter_y[i]})", (scatter_x[i], scatter_y[i]), 
                        textcoords="offset points", xytext=(5,5), ha='right')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
axs[1, 0].grid(True, linestyle="--", alpha=0.6)
axs[1, 0].legend()

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, shadow=True)
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
